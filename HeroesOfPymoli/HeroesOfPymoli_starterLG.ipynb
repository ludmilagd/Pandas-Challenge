```python
# Dependencies and Setup- LUDMILA HOMEWORK
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
total_players= len(purchase_data_df["SN"].unique())
total_players
total_sales=purchase_data_df["SN"].count()
total_sales
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items =purchase_data_df["Item ID"].unique()
unique_items=len(unique_items)
unique_items

average_price=purchase_data_df["Price"].mean()
average_price

total_purchase=purchase_data_df["Price"].sum()
total_purchase

count_purchase=purchase_data_df["Price"].count()
count_purchase

summary_table=pd.DataFrame({"Number of unique items":unique_items,
                           "Average Price":[average_price],
                           "Number of purchases":[count_purchase],
                           "Total revenue":[total_purchase]})


summary_table["Average Price"]=summary_table["Average Price"].map("${:,.2f}".format)
summary_table["Total revenue"]=summary_table["Total revenue"].map("${:,.2f}".format)

summary_table
```
